```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn import neighbors
from sklearn.metrics import mean_squared_error 
import matplotlib.pyplot as plt
from numpy import where
%matplotlib inline
data1 =pd.read_csv("diabetes.csv") #data cleaning
data1['Outcome']=data1['Outcome'].replace('negative',0)
data1['Outcome']=data1['Outcome'].replace('positive',1)
X = data1.drop(columns=['Outcome'])
y = data1['Outcome'].values
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=60,test_size=0.3, stratify=y)#splitting the data
model = KNeighborsClassifier(n_neighbors = 3)#building the model
# Fit the classifier to the data
model.fit(X_train,y_train)
print("Predicted Target Variable Values Are=")
pred=model.predict(X_test)#prediction
print(pred)
print("Accuracy of knn model Designed:%f"%(model.score(X_test, y_test)))
####
range_k=range(1,30)
k_values=[]
mse_val = [] #to store mse values for different k
for i in range_k:
    i = i+1
    model = neighbors.KNeighborsClassifier(n_neighbors = i)
    model.fit(X_train, y_train)  #fit the model
    pred=model.predict(X_test) #make prediction on test set
    error = mean_squared_error(y_test,pred) #calculate mse
    mse_val.append(error) #store mse values
    k_values.append(i) 
print('Minimum MSE or Loss is:',(error.min()))#Loss value
plt.figure() 
plt.plot(k_values,mse_val, color ='blue', linestyle ='dashed', marker ='o', markerfacecolor ='red', markersize = 3)
plt.title('Error Rate vs. K Value') 
plt.xlabel('K value in Knn') 
plt.ylabel('Error Rate in Knn for each k') 
#cross validation check
all_means=[]
for i in range_k:
    knn_cv = KNeighborsClassifier(n_neighbors=i)
    cv_scores = cross_val_score(knn_cv, X, y, cv=20)
    all_means.append(np.mean(cv_scores))
print("Cross validation accuracy:%f"%(max(all_means)))
print("Efficient k-value according to cross validation method:%d"%(all_means.index(max(all_means))+1))#As we took k value from index so add 1
#plot
plt.figure()
plt.plot(range_k, all_means)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-validated accuracy')


    

```
